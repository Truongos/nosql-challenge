## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question: 
* Use `count_documents` to display the number of documents contained in the result.
* Display the first document in the results using `pprint`.

### 1. Which establishments have a hygiene score equal to 20?

```python
# Query for establishments with a hygiene score of 20 within the "scores" field
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents contained in the result
count = establishments.count_documents(query)
print(f"Number of documents with a hygiene score of 20: {count}")

# Display the first document in the results using pprint
result = establishments.find(query)
first_document = result[0] if count > 0 else None
if first_document:
    pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(list(result))

# Display the number of rows in the DataFrame
num_rows = len(df)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
if num_rows > 0:
    print(df.head(10))
else:
    print("No documents with a hygiene score of 20 were found.")
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex':'London'}, 'RatingValue': {'$gte': 4}}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of documents in London with RatingValue >= 4: {count}")

# Display the first document in the results using pprint
result = establishments.find(query)
first_document = result[0] if count > 0 else None
if first_document:
    pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(list(result))

# Display the number of rows in the DataFrame
num_rows = len(df)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
if num_rows > 0:
    print(df.head(10))
else:
    print("No documents matching the criteria were found.")
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
query = {'BusinessName': 'Penang Flavours'}


penang = establishments.find_one(query)

degree_search = 0.01
latitude = penang['geocode']['latitude']
longitude = penang['geocode']['longitude']

print(latitude)
print(longitude)

query = {
     'geocode.latitude': {
         '$gte': latitude - degree_search,
         '$lte': latitude + degree_search
     },
    'geocode.longitude': {
        '$gte': longitude - degree_search,
        '$lte': longitude + degree_search
    },
   'RatingValue': 5
}

sort =  [('scores.Hygiene', -1)]

results = list(establishments.find(query).sort(sort).limit(5))

# Print the results
pprint(results)
```

```python
# Convert result to Pandas DataFrame
df = pd.DataFrame(results)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
match = {'$match': {'scores.Hygiene': 0}}
group = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum':1}}}
sort = {'$sort': {'count': -1}}

pipeline = [match, group, sort]

# Print the number of documents in the result
results = list(establishments.aggregate(pipeline))
# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(len(df))
# Display the first 10 rows of the DataFrame
df.head(10)
```
